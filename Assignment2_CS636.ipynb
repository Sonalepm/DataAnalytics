```python
import pandas as pd
import csv
```

```python
def find_divident(file):
    data = pd.read_csv(file)
    df= pd.DataFrame(data)
    #print(df)
    Adj_close_ratio_list = [0]
    Close_ratio_list = [0]
    date_list = []
    divident_list = []
    for i in range (1,len(df)):
        Adj_close_ratio_list.append(round(df['Adj Close'][i-1]/df['Adj Close'][i], 3))
        Close_ratio_list.append(round(df['Close'][i-1]/df['Close'][i], 3))
    df['Close Ratio'] = Close_ratio_list
    df['Adj Close Ratio'] =  Adj_close_ratio_list
    print(df)
    for i in range(len(df)-1):
        if (df['Close Ratio'][i] != df['Adj Close Ratio'][i]) :
            date_list.append(df['Date'][i])
            divident_list.append(((df['Close Ratio'][i] - df['Adj Close Ratio'][i] )* df['Close'][i]))
    final_df = pd.DataFrame(list(zip(date_list,divident_list)), columns = ['Date','Divident'])
    return (final_df)
```
